```
class Calculator:
  def __init__(self):
    self.value = 0
    
  def add(self, n):
    self.value = self.value + n
    print(self.value)

  def substract(self, n):
    self.value = self.value - n
    print(self.value)

  def multiply(self, n):
    self.value = self.value * n
    print(self.value)

  def divide(self, n):
    if n == 0 :
      print('You can not divide with 0')
      return
    else:
      self.value = self.value/n
      print(self.value)
```

```
a = Calculator()
a.add(2)
a.substract(1)
a.multiply(5)
a.divide(5)
```
